### Que 12 Solution

```python
# Importing necessary libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_T2 = np.array(data['T2'])
y_T6 = np.array(data['T6'])
x_T2 = x_T2.reshape(-1, 1)

train_x, test_x, train_y, test_y = train_test_split(x_T2, y_T6, train_size = 0.7, random_state = 42)

linear = LinearRegression()
linear.fit(train_x, train_y)
prediction_y = linear.predict(test_x)

from sklearn.metrics import r2_score
r_squared = r2_score(test_y, prediction_y)
round(r_squared, 2)
```

### Que 13 Solution
Finding the Mean Absolute Error

### Using MinMaxScaler to normalize the data

```python
from sklearn.preprocessing import MinMaxScaler
minMaxScaler = MinMaxScaler()
data_normalize = pd.DataFrame(minMaxScaler.fit_transform(data), columns = data.columns)
```

```python
data_normalize.head()
```

```python
data_normalize.tail()
```

### Separating the dataset into predictive features and target
Taking Appliance as the target feature

```python
X = data_normalize.drop(['Appliances'], axis = 1)
y = data_normalize['Appliances']
```

### Splitting the dataset into train and test sets using train_test_split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 42)
```

```python
linear = LinearRegression()
linear.fit(X_train, y_train)
y_prediction = linear.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_prediction)
print(round(mae, 2))
```

### Que 14 Solution

```python
## Residual Sum of Squares(RSS)

residual_ss = np.sum(np.square(y_test - y_prediction))
print(round(residual_ss, 2))
```

### Que 15 Solution

```python
## Root Mean Squared Error(RMSE)

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_prediction))
print(round(rmse, 3))
```

### Que 16 Solution

```python
from sklearn.metrics import r2_score
r_squared = r2_score(y_test, y_prediction)
round(r_squared, 2)
```

### Que 17 Solution

```python
#to check the weights of the features using linear regression

def weights(model, feature, column):
    #this function returns the weight of every feature
    weights = pd.Series(linear.coef_, feature.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', column]
    weights_df[column].round(3)
    return weights_df

linear_weights = weights(linear, X_train, 'Linear_model_Weight')
linear_weights
```

```python
print(linear_weights['Linear_model_Weight'].min())
print(linear_weights['Linear_model_Weight'].max())
```
